```python
import os
import numpy as np
import pandas as pd

import seaborn as sns
import plotly.express as px
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.manifold import TSNE
from sklearn.decomposition import PCA
from sklearn.metrics import euclidean_distances
from scipy.spatial.distance import cdist
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
import warnings
warnings.filterwarnings("ignore")
```

```python
data = pd.read_csv("/content/ndatsetwithgenres.csv")
```

```python
from sklearn.preprocessing import OrdinalEncoder
from sklearn.impute import SimpleImputer
# Select the useful columns as features
data = data[:5000]
X = data[['artist_name', 'Genre']].astype('category')
print(type(X))
print(len(X))
Column = Pipeline([('OE',OrdinalEncoder()),('Imputer',SimpleImputer())])
X_new = Column.fit_transform(X)
# Target variable
y = data['unique_id']
```

```python
cluster_pipeline = Pipeline([('scaler', StandardScaler()), ('kmeans', KMeans(n_clusters=10))])
cluster_pipeline.fit(X_new)
data['cluster'] = cluster_pipeline.predict(X_new)
```

```python
from sklearn.manifold import TSNE

tsne_pipeline = Pipeline([('scaler', StandardScaler()), ('tsne', TSNE(n_components=2, verbose=1))])
embedding = tsne_pipeline.fit_transform(X_new)
projection = pd.DataFrame(columns=['x', 'y'], data=embedding)
projection['genres'] = data['Genre']
projection['cluster'] = data['cluster']

fig = px.scatter(
    projection, x='x', y='y', color='cluster', hover_data=['x', 'y', 'genres'])
fig.show()
```

```python
song_cluster_pipeline = Pipeline([('scaler', StandardScaler()),
                                  ('kmeans', KMeans(n_clusters=20,
                                   verbose=False))
                                 ], verbose=False)

X = data.select_dtypes(np.number)
number_cols = list(X.columns)
song_cluster_pipeline.fit(X)
song_cluster_labels = song_cluster_pipeline.predict(X)
data['cluster_label'] = song_cluster_labels
```

```python
from sklearn.decomposition import PCA

pca_pipeline = Pipeline([('scaler', StandardScaler()), ('PCA', PCA(n_components=2))])
song_embedding = pca_pipeline.fit_transform(X)
projection = pd.DataFrame(columns=['x', 'y'], data=song_embedding)
projection['title'] = data['track_name']
projection['cluster'] = data['cluster_label']
fig = px.scatter(
    projection, x='x', y='y', color='cluster', hover_data=['x', 'y', 'title'])
fig.show()
```

```python
def recommend_songs(input_songs, data):
    """
    Function to recommend songs based on input songs using clustering results.
    """
    input_clusters = []
    for song in input_songs:
        genre_value = data[data['track_name'] == song['track_name']]['Genre'].values
        for i in range(len(genre_value)):
            cluster = data[data['Genre'] == genre_value[i]]['cluster_label'].values
            if len(cluster) > 0:
                input_clusters.extend(cluster)

    # Check if input_clusters is not empty
    if not input_clusters:
        return "No recommendation available, input songs not found in the dataset."

    # Get the most common cluster among input songs
    recommended_cluster = max(set(input_clusters), key=input_clusters.count)

    # Filter songs in the recommended cluster
    recommended_songs = data[data['cluster_label'] == recommended_cluster]

    # Check if recommended_songs is not empty
    if recommended_songs.empty:
        return "No recommendation available for the given input songs and cluster."

    # Get 40 random songs from the recommended cluster
    recommended_songs_sample = recommended_songs.sample(40)

    return recommended_songs_sample

# Example usage
input_songs = []
for i in range(5):
    song_dict = {}
    track_name = input("Enter track name:")
    artist_name = input('Enter Artist name:')
    song_dict['track_name'] = track_name
    song_dict['artist_name'] = artist_name
    input_songs.append(song_dict)

recommended_tracks = recommend_songs(input_songs, data)
if isinstance(recommended_tracks, str):
    print(recommended_tracks)
else:
    print("Recommended Songs:")
    for index, row in recommended_tracks.iterrows():
        print(f"{row['track_name']} in Cluster {row['cluster_label']}")
```
